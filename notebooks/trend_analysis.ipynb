```python
import pytrends
import pandas as pd
from pytrends.request import TrendReq
# from pytrends.request import g
import altair as alt
```

```python
pytrend = TrendReq(hl='india')
```

### Rana Daggubati

```python
pytrend.build_payload(kw_list=["Rana Daggubati"])
```

```python
pytrend.suggestions("Rana")
```

```python
relatedDF=pytrend.related_topics()
```

```python
relatedqueriesDF=pytrend.related_queries()
```

```python
relatedqueriesDF['/m/09ggjsc']['rising']
```

```python
relatedqueriesDF['/m/09ggjsc']['top']
```

```python
interest_by_regionDF=pytrend.interest_by_region()
```

```python
interest_by_timeDF=pytrend.interest_over_time()
```

```python
chart=alt.Chart(interest_by_regionDF.sort_values(by="Rana Daggubati",ascending=False).reset_index())
```

```python
chart.mark_line().encode(alt.X("geoName"),alt.Y("Rana Daggubati",title="Freq of search"))
```

```python
interest_by_timeDF1=interest_by_timeDF.reset_index()
time_chart=alt.Chart(interest_by_timeDF1,width=1000)
```

```python
interest_by_timeDF1
time_chart.mark_line().encode(alt.X('date'),alt.Y('/m/09ggjsc',title='Freq of search')).interactive()
```
