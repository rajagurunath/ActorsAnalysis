```python
consumer_key=""
consumer_secret=""
```

```python
auth = tweepy.AppAuthHandler(consumer_key, consumer_secret)
api = tweepy.API(auth)

```

```python
user = api.get_user('@RanaDaggubati')
print(user.screen_name)
print(user.followers_count)
# for friend in user.friends():
#    print(friend.screen_name)
```

```python
frds=user.friends()
```

```python
frdsDF=pd.json_normalize([f._json for f in frds])
frdsDF.head()
```

```python
frdsDF.columns.tolist()
```

```python
chart = alt.Chart(frdsDF)
```

```python
chart.mark_point().encode(x="followers_count",y="friends_count",
                          tooltip=["followers_count","friends_count",'location'],color='location').interactive()
```

```python
selection = alt.selection_multi(fields=['location'])
color = alt.condition(selection,
                      alt.Color('location:N', legend=None),
                      alt.value('lightgray'))

scatter = alt.Chart(frdsDF).mark_bar().encode(
    x='followers_count',
    y='friends_count',
    color=color,
    tooltip=['location','followers_count']
).interactive()

legend = alt.Chart(frdsDF).mark_point().encode(
    y=alt.Y('location', axis=alt.Axis(orient='right')),
    color=color
).add_selection(
    selection
)

(scatter | legend)
```
