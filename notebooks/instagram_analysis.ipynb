```python
text=[]
jpg=[]

for f in text_files:
    with open(f,'rb') as file:
        _text=file.read().decode()
    
    tmpJpeg=glob.glob(r"C:\Users\GLANKUP\Guru\rakulpreet\{}*.jpg".format(f.split(os.sep)[-1].split('.')[0]))
    text.extend([_text]*len(tmpJpeg))
    jpg.extend(tmpJpeg)
    print(len(tmpJpeg),len(text))
                      
```
